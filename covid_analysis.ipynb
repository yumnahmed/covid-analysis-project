## Handle Missing Values

```python
data.drop(['NewCases', 'NewDeaths', 'NewRecovered'], axis=1, inplace=True)
```

```python
data["Continent"]=data["Continent"].fillna(data["Continent"].mode()[0])
data["WHO Region"]=data["WHO Region"].fillna("unknowen")
```

```python
data["Population"]=data["Population"].fillna(data["Population"].median())
data["TotalDeaths"]=data["TotalDeaths"].fillna(0)
data["TotalRecovered"]=data["TotalRecovered"].fillna(0)
data["ActiveCases"]=data["ActiveCases"].fillna(0)
data["Serious,Critical"]=data["Serious,Critical"].fillna(0)
data["TotalTests"]=data["TotalTests"].fillna(0)
```

```python
data.isnull().sum()
```

```python
data['Tot Cases/1M pop'] = (data['TotalCases'] / data['Population']) * 1000000
data['Deaths/1M pop'] = (data['TotalDeaths'] / data['Population']) * 1000000
data['Tests/1M pop'] = (data['TotalTests'] / data['Population']) * 1000000
```

```python
data.isnull().sum()
```

## Add New Features

```python
data['DeathRate'] = data['TotalDeaths'] / data['TotalCases']
data['RecoveryRate'] = data['TotalRecovered'] / data['TotalCases']
data['ActiveRate'] = data['ActiveCases'] / data['TotalCases']
data['TestsRate'] = data['TotalDeaths'] / data['TotalCases']
```

## Check Data Types

```python
data.dtypes
```

```python
data["Population"]=data["Population"].astype(int)
data["TotalCases"]=data["TotalCases"].astype(int)
data["TotalDeaths"]=data["TotalDeaths"].astype(int)
data["TotalRecovered"]=data["TotalRecovered"].astype(int)
data["ActiveCases"]=data["ActiveCases"].astype(int)
data["Serious,Critical"]=data["Serious,Critical"].astype(int)
data["Tot Cases/1M pop"]=data["Tot Cases/1M pop"].astype(int)
data["Deaths/1M pop"]=data["Deaths/1M pop"].astype(int)
data["TotalTests"]=data["TotalTests"].astype(int)
data["Tests/1M pop"]=data["Tests/1M pop"].astype(int)
```
